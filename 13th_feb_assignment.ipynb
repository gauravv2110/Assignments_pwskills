ArithmeticError class exceptions :

This class is the base class for those built-in exceptions that are raised for various arithmetic errors such as :

1. OverflowError
2. ZeroDivisionError
3. FloatingPointError

* 1. ZeroDivisionError 

```python
try : 
    
    a = 55 / 0

except ZeroDivisionError as e :
    
    print(f"this is the error caused because of {e}")
```

* 2. FloatingPointError 

```python
try:
    j = 5.0
    for i in range(1, 1000):
        j = j**i
        print(j)
except OverflowError as e:
    print(f"{e} Overflow error happened result is too largee")
```

LookupError exceptions:

This is the base class for those exceptions that are raised when a key or index used on a mapping or sequence is invalid or not found. The exceptions raised are :

1. KeyError
2. IndexError

* 1. KeyError 

```python
try :
    
    d = {1:'gaurav' , 2:'patokar' , 4:'shivraj' , 5:'patokar'}
    print(d[3])

except KeyError as e :
    
    print(f"{e} , because there is no given key in the dictionary availabel")
    
    
```

* 2. IndexError

```python
try :
    
    l = [1,2,3,4,5]
    print(l[5])

except IndexError as e :
    
    print(f"{e} , because our list only contains 4 indexes")
```
